```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.circuit.library import DiagonalGate
import numpy as np
cr1 = ClassicalRegister(7)
qr = QuantumRegister(7)
qc = QuantumCircuit(qr, cr1)
for i in range(0,7):
    qc.h(i)
    qc.barrier(i)
array = [1,-1]
for i in range(126):
    array.append(1) 
qc.append(DiagonalGate(array), [0,1,2,3,4,5,6])
for i in range(0,7):
    qc.h(i)
array2 = [1]
for j in range(127):
    array2.append(-1)
qc.append(DiagonalGate(array2), [0,1,2,3,4,5,6])
for i in range(0,7):
    qc.h(i)
    qc.barrier(i)

qc.measure([0,1,2,3,4,5,6],cr1)
qc.draw(output='mpl')
```
